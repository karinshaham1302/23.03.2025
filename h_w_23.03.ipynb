```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
mobile = pd.read_csv("mobile.csv")
df_mobile = mobile.copy()
```

```python
df_mobile.info()
```

```python
df_mobile.describe().transpose()
```

```python
# 1.
df_mobile['First Name'] = df_mobile['Customer Name'].apply(lambda x: str(x)[:5])
display(df_mobile[['First Name', 'Customer Name']].head())

# ?
```

```python
# 2.
df_mobile['phone_company'] = df_mobile['Phone Number'].replace({'050': 'Pelephone', '051': 'Golan', '052': 'Cellcom', '053': 'Motorola'})
df_mobile[['Phone Number', 'phone_company']].head()

# ?
```

```python
# 3.
# a.
display(df_mobile['phone_company'].value_counts().head(10))
print()

# b.
display(df_mobile['phone_company'].unique())
print()

# c.
display(df_mobile['phone_company'].nunique())
```

```python
# 4.
# a.
df_mobile['Type'] = df_mobile.apply(lambda row: 'talker' if row['Talked'] > row['Internet'] else 'surfer', axis=1)
display(df_mobile[['Customer Name', 'Talked', 'Internet', 'Type']].tail(10))


# b.
def talker_type(talker, surfer):
    if talked > surfer:
        return 'talker'
    else:
        return 'surfer'

df_mobile['talker_type'] = np.vectorize(talker_type)(df_mobile['Talked'], df_mobile['Internet'])
display(df_mobile[['Customer Name', 'Talked', 'Internet', 'talker_type']].head(10))
```

```python
# 5.
df_sorted = df_mobile.sort_values(by='Talked', ascending=True)
display(df_sorted[['Customer Name', 'Talked']].head(10))


df_mobile['Time'] = df_mobile['Talked'] + df_mobile['Internet']
df_sorted = df_mobile.sort_values(by='Time', ascending=True)
display(df_sorted[['Customer Name', 'Talked', 'Internet', 'Time']].head(10))
```

```python
# 6.
display(df_mobile.nlargest( 5 , 'Talked'))

display(df_mobile.nsmallest( 3 , 'Internet'))
```
